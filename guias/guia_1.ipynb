### Generar fechas aleatorias

```python
def random_dates(start, end, n=10):

    start_u = start.value//10**9
    end_u = end.value//10**9

    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')
```

## Ejercicio 4

4- Un sitio de Ebooks tiene información sobre los reviews que los usuarios hacen de sus libros en un DataFrame con formato (user_id, book_id, rating, timestamp). Por otro lado tenemos información en otro DataFrame que bajamos de GoodReads: (book_id, book_name,
avg_rating). Podemos suponer que los Ids de los libros son compatibles. Se pide usar Python
Pandas para:
* A. Obtener un DataFrame que indique el TOP5 de Ebooks en el sitio de Ebooks. (Para este punto se puede ignorar el segundo DataFrame).
* B. Obtener un DataFrame que indique qué libros tienen una diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads.

### Set de prueba

```python
df_ebooks = pd.DataFrame(columns = ['user_id', 'book_id', 'rating', 'timestamp'])
df_goodreads = pd.DataFrame(columns = ['book_id', 'book_name', 'avg_rating'])
libros = ['Narnia', 'Mi Queso', 'Jesus', 'Data Mining', 'Colores']
for id_libro in range(len(libros)) :
    for j in range(3):
        df_ebooks = df_ebooks.append({
                        'user_id' : j + 200,
                        'book_id' : id_libro,
                        'rating' : random.uniform(0, 100),
                        'timestamp' : random.choice(['2017-8-1','2018-1-2','2019-9-4','2019-9-9'])
                    }, ignore_index = True)
    df_goodreads = df_goodreads.append({
                            'book_id' : id_libro,
                            'book_name' : libros[id_libro],
                            'avg_rating' : random.uniform(0, 100)
                        }, ignore_index = True)   
```

```python
df_ebooks
```

```python
df_goodreads
```

### Resolucion

#### A. Obtener un DataFrame que indique el TOP5 de Ebooks en el sitio de Ebooks. (Para este punto se puede ignorar el segundo DataFrame).

```python
df_ebooks_avg_rating =  df_ebooks.groupby(['book_id']).agg({'rating' : 'mean'}).nlargest(5, 'rating')
df_ebooks_avg_rating.columns = ['avg_rating_ebooks']
df_ebooks_avg_rating = df_ebooks_avg_rating.reset_index()
df_ebooks_avg_rating
```

#### B. Obtener un DataFrame que indique qué libros tienen una diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads.

```python
df_4_merge = pd.merge(df_ebooks_avg_rating, df_goodreads, on = 'book_id', how = 'inner')
df_4_merge['diferencia_avg_rating'] = abs(df_4_merge['avg_rating_ebooks'] - df_4_merge['avg_rating'])
df_4_merge['diferencia_mayor_20_porciento'] = df_4_merge['diferencia_avg_rating'] > 20
df_4_merge
```

#### B. Indicar el top-10 de productos que se vendieron mayor cantidad de días de forma consecutiva.

```python
def max_dias_consecutivos(df):
    dias_consecutivos_max = 0
    dias_consecutivos  = 0
    fecha_anterior = df['fecha'].iloc[0] 
    for i in range(1,df['fecha'].size):
        fecha = df['fecha'].iloc[i]
        delta_dias = (fecha - fecha_anterior) / np.timedelta64(1,'D')
        if (delta_dias == 1):
            dias_consecutivos += 1
            print(dias_consecutivos)
        else:
            if (dias_consecutivos > dias_consecutivos_max):
                dias_consecutivos_max = dias_consecutivos
            dias_consecutivos = 0
        fecha_anterior = fecha
    return dias_consecutivos_max
```
